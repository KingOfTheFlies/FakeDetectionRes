```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob

# Считываем все CSV файлы
# csv_files = glob.glob("./NPR-DeepfakeDetection/results/*.csv")
csv_files = ['./NPR-DeepfakeDetection/results/my_train_model_epoch_0.csv',
             './NPR-DeepfakeDetection/results/my_train_model_epoch_20.csv',
            './NPR-DeepfakeDetection/results/my_train_model_epoch_40.csv',
            './NPR-DeepfakeDetection/results/my_train_model_epoch_60.csv',
            './NPR-DeepfakeDetection/results/my_train_model_epoch_80.csv',
            './NPR-DeepfakeDetection/results/my_train_model_epoch_98_epoch.csv',
             ]

# Создаем пустой DataFrame для объединения всех данных
combined_data = pd.DataFrame()

# Проходим по всем CSV файлам и добавляем информацию в общий DataFrame
for file in csv_files:
    model_name = file.split("/")[-1].replace(".csv", "")  # Получаем имя модели из имени файла
    df = pd.read_csv(file)
    df['model'] = model_name  # Добавляем столбец с именем модели
    combined_data = pd.concat([combined_data, df], ignore_index=True)

# Получаем уникальные группы
groups = combined_data['group'].unique()

# Построение графиков для accuracy
plt.figure(figsize=(14, len(groups) * 4))
for i, group in enumerate(groups):
    plt.subplot(len(groups), 1, i + 1)
    sns.barplot(x='dataset', y='accuracy', hue='model', data=combined_data[combined_data['group'] == group])
    plt.title(f'Comparison of Accuracy for Group {group}')
    plt.ylabel('Accuracy')
    plt.xticks(rotation=45, ha="right")
    plt.legend(title='Model')

plt.tight_layout()
plt.show()

# Построение графиков для average_precision
plt.figure(figsize=(14, len(groups) * 4))
for i, group in enumerate(groups):
    plt.subplot(len(groups), 1, i + 1)
    sns.barplot(x='dataset', y='average_precision', hue='model', data=combined_data[combined_data['group'] == group])
    plt.title(f'Comparison of Average Precision for Group {group}')
    plt.ylabel('Average Precision')
    plt.xticks(rotation=45, ha="right")
    plt.legend(title='Model')

plt.tight_layout()
plt.show()
```
